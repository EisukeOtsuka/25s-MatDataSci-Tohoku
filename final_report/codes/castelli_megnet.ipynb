```python
from matminer.datasets import load_dataset
import pandas as pd

# データロード
df = load_dataset("castelli_perovskites")
df = df[df["gap gllbsc"].notnull()].reset_index(drop=True)

# 重要: pymatgenの構造データとバンドギャップだけを使う
structures = df["structure"].tolist()
targets = df["gap gllbsc"].values
```

```python
from megnet.models import MEGNetModel

# ガウス展開の中心と幅を設定
centers = list(range(0, 5))     # 例えば0Å〜4Åまで1Å刻み
width = 0.5

model = MEGNetModel(
    nfeat_bond=100,
    nfeat_global=2,
    nfeat_node=100,
    ntarget=1,
    loss="mae",
    metrics=["mae"],
    verbose=1,
    centers=centers,        # ← 正しい引数名
    width=width             # ← 正しい引数名
)
```
